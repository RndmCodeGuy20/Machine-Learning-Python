```python
import numpy as np
from scipy import stats as st
import matplotlib.pyplot as plt
```

## Mean

```python
xData = [99,86,87,88,111,86,103,87,94,78,77,85,86]

xMean = np.mean(xData)

xMean
```

## Median

```python
xMed = np.median(xData)

xMed
```

## Mode

```python
xMode = st.mode(xData)

xMode
```

## Standard Deviation

```python
xStd = np.std(xData)

xStd
```

## Variance

```python
xVar = np.var(xData)

xVar
```

## Percentile

```python
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

xPer = np.percentile(ages, 100)

print(xPer)
```

## Random Uniform Distribution

```python
xPer = np.percentile(xRand, 91.04)

xPer
```

```python
colors = ['#003f5c', '#58508d', '#bc5090', '#ff6361', '#ffa600']
```

## Histogram

```python
xHist = np.random.uniform(-60, 300, 1000000)

plt.figure(figsize=(20,10))

plt.hist(xHist, 300, color=colors[2])
plt.show()
```

```python
xHist = np.random.normal(-60, 300, 1000000)

plt.figure(figsize=(20,10))

plt.hist(xHist, 300,color=colors[3])
plt.show()
```
