> Assumption : slope of the line was found to be 0.2027 (using linear regression)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('darkgrid')
colors = ["#ffbe0b", "#fb5607", "#ff006e", "#8338ec", "#3a86ff",]
font = {'family': 'Arial',
        'color':  colors[1],
        'weight': 'normal',
        'size': 16,
        }
```

```python
weightVals = [4.53, 0.53, 5.05]
heightVals = [2.19, 2.63, 4.41]
```

```python
plt.figure(figsize=(10,7.5))

plt.scatter(weightVals, heightVals, color='#fff', alpha=0.7, edgecolors=colors[2], s=50)

plt.xlabel("Weight", fontdict=font)
plt.ylabel("Height", fontdict=font)

plt.show()
```

```python
weight = np.array(weightVals)
height = 0.2027*weight + 2.3935

plt.figure(figsize=(10,7.5))

plt.plot(weight, height, color=colors[0])
plt.scatter( weightVals, heightVals, color='#fff', alpha=0.7, edgecolors=colors[2], s=50)

plt.xlabel("Weight", fontdict=font)
plt.ylabel("Height", fontdict=font)
plt.title("Actual Regression Line")

plt.show()
```

```python
# Plotting the gradient descent plot
c = 0.0
plt.figure(figsize=(10,7.5))

while c <= 5:
    sumSquared = 0
    for i in range(3):
        predHeight = 0.2027*weightVals[i] + c
        sumSquared += (heightVals[i] - predHeight)**2
    plt.scatter(c, sumSquared, color=colors[1], s=10)
    c += 0.1
```
